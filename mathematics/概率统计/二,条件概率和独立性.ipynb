## 1, 条件概率

例1： 考虑恰好有两个小孩的家庭,并且假定生男, 生女是等可能的。如果从这些家庭中随机地选择一个孩子，并发现她为女孩，问在她家里另一个孩子是男孩的概率是多少？

解:
   样本空间: {男g,男b,女g,女b}，
   
   设事件A为 "这个孩子是女孩"， 事件B为"这个孩子有一个兄弟"
   
   A={女g，女b}， B={男b， 女b}
   
   AB = {女b}
   
   所求概率为 p（B|A） = $\frac{P(AB)}{P(A)}$= $\frac{ \frac{1}{4}}{\frac{2}{4}}$ = $\frac{1}{2}$

## 2, 有关条件概率的三个重要计算公式

### ① 乘法公式

1. 设A,B是两个事件,P(B) > 0, 则P(AB) = P(B)P(A | B)

    证明: P(A | B) = $\frac{P(AB)}{P(B)}$  $\Rightarrow$ P(AB) = P(B)P(A|B)

2. 设$A_1,A_2,A_3,...A_n$为n个事件,且$P(A_1, A_2,...A_{n-1})>0$，则

$P（A_1, A_2, ... A_n）= P(A_1) * P(A_2 | A_1) * P(A_3 | A_1 A_2)...P(A_n | A_1 A_2 ... A_{n-1})$

   证明: 数学归纳法,设 $P(A_1...A_k) = P(A_1) * P(A_2 | A_1) * P(A_k | A_1...A_{k-1})$,
    
   $P(A_1 ... A_{k+1}) = P(A_1 A_2 ... A_k)* P(A_{k+1} | A_1 A_2 ... A_k)$
    
   = $P(A_1)*P(A_2|A_1)...P(A_{k+1} | A_1 A_2 ... A_k)$


直接验证

$P(A_1)*P(A_2|A_1)*P(A_3 | A_1 A_2) ... P(A_n | A_1 A_2 ... A_{n-1})$

= $P(A_1)\frac{P(A_1 A_2)}{P(A_1)} * \frac{P(A_1 A_2 A_3)}{P(A_1 A_2)} ... \frac{P(A_1 A_2 ... A_n)}{P(A_1 A_2 ... A_{n-1})} $

= $P(A_1 A_2 ... A_n)$


例1:

   设箱子内有a个白球,b个黑球,在其中不放回地连取3次,问前2次取到白球第3次取到黑球的概率。
   
   解: 设事件$A_i$ 表示第i次抽到白球，
   
   $P(A_1 A_2 \bar{A_3}) = P(A_1) * P(A_2 | A_1) * P(\bar{A_3} | A_1 A_2)$
   
  = $\frac{a}{a+b}*\frac{a-1}{a+b-1}*\frac{b}{a+b+2}$


### ② 全概率公式

设 $B_1, B_2 ... B_n$ 是样本空间 Ω 的一个分割,即 $B_1, B_2, ... B_n$ 互不相容,且$$ \quad \bigcup_{i=1}^{n} B_i = $$Ω 。如果P(B_i) > 0 (i = 1, 2, ..., n), 则对任一事件A，有

$$P(A) = \sum_{i=1}^n P(B_i) P(A | B_i)$$

证明： $P(A) = P(AΩ) = P(A\bigcap(B_1 \bigcup B_2 \bigcup ... \bigcup B_n))$

$$=P(AB_1 \bigcup AB_2 \bigcup ... \bigcup AB_n)$$
$$=P(AB_1) + P(AB_2) + ... + P(AB_n)$$
$$=P(B_1)P(A|B_1)+ P(B_2)P(A|B_2) + ... + P(B_n)P(A|B_n)$$

例1: 

 设甲箱中有a个白球，b个黑球，a > 0, b > 0, 乙箱中有c个白球，d个黑球。自甲箱中任取一球放入乙箱，然后再从乙箱中任取一个球。求最后由乙箱取出的是白球的概率。
 
  解: 设事件A表示最后由乙箱取出的是白球,事件W表示从甲箱取出白球，
  
  $P(A) = P(W)P(A|W) + P(\bar{W})P(A|\bar{W})$
  
  $$= \frac{a}{a+b}*\frac{c+1}{c+d+1}+\frac{b}{a+b}*\frac{c}{c+d+1}$$

  $$= \frac{ac + bc +1}{(a + b)(c + d +1)}$$
  
例2:

买彩票,设n张彩票中有1张奖券，人们排成一队购买彩票，求第k个人购到奖券的概率。

解1: 手部分析法 设置$A_k(n)$为事件 "n个人买彩票, 第k个人中奖", 则

$P(A_k(n)) = P(A_1(n))P(A_k(n) | A_1(n)) + P(\bar{A_1(n)})P(A_k(n) | \bar{A_1(n)})$

$$= 0+ \frac{n - 1}{n} P(A_{k-1})$$

$$= \frac{n-1}{n} \frac{n-2}{n-1}P(A_{k-2}(n-2))$$

$$= \frac{n-1}{n} \frac{n-2}{n-1}... \frac{n-(k-1)}{n-(k-2)}P(A_1(n-(k-1)))$$

$$= \frac{n-1}{n}\frac{n-2}{n-1}...\frac{n-(k-1)}{n-(k-2)}\frac{1}{n-(k-1)} = \frac{1}{n}$$

解2:

 设事件$A_i$表示事件"第i个人中奖",则
 
 $P(A_k) = P(\bar{A_1}\bar{A_2}...\bar{A_{k-1}}A_k)$
 
 $$= P(\bar{A_1})P(\bar{A_2} | \bar{A_1})...P(\bar{A_{k-1} | \bar{A_1} \bar{A_2} ... \bar{A_{k-2}}})P(A_k|\bar{A_1}\bar{A_2}...\bar{A_{k-1}})$$
 
 $$= \frac{n-1}{n} \frac{n-2}{n-1}...\frac{n-k+1}{n-k+2}\frac{1}{n-k+1} = \frac{1}{n}$$


### ③ 贝叶斯(Bayes)公式

设$B_1, B_2, ... B_n$, 是样本空间Ω的一个分割,即$B_1, B_2，... B_n$ 互不相容，且$$ \quad \bigcup_{i=1}^{n} B_i = $$Ω,如果 $P(A) > 0, P(B) > 0, (i = 1, 2 ... n)$,

则 $P(B_i | A) = \frac{P(B_i)P(A | B_i)}{\sum_{i=1}^n P(B_i)  P(A|B_i)}$

证明: $P(P(B_i | A)) = \frac{P(A | B_i)}{P(A)}$

$$= \frac{P(B_i)P(A | B_i)}{P(B_i)P(A|B_1)+P(B)P(A|B_2)+...+P(B_n)(A|B_n)}$$

例1:

某考生回答一道有4个选择题，设会答该题的管理是P,并且会答时一定能答对，若不会时则在4个答案中任选择1个。求该考生回答正确时他确实会答的概率。

解: 设事件A表示"答对",事件B表示"会答",则

$$P(B | A) = \frac{P(B)P(A|B)}{P(B)P(A|B) + P(B)P(A|\bar{B})}$$

$$= \frac{p*1}{p*1 + (1 -p) * \frac{1}{4}}$$

$$= \frac{4p}{1+3p}$$

例2:

  一地区某疾病的发病率是0.0004.现有一种化验方法，真正得病得人，其化验结果99%呈阳性，对未患病者，化验结果99.9%呈阴性。求下列两事件得发生概率；
  
  1. 检查结果呈阳性，是否真的患病？
  2. 检查结果呈阴性，是否就可以放心地认为自己没有病？
  
解: 设事件A表示事件"化验呈阳性"，B表示"患病",则检查结果呈阳性，但实际上没有患病(虚惊一场的概率)

 $$P(\bar{B} | A) = \frac{P(\bar{B}P(A | \bar{B}))}{P(B)P(A|B) + P(\bar{B}P(A|\bar{B}))} = \frac{0.9996 * 0.001}{0.0004 * 0.99 + 0.9996 * 0.001} = 0.716$$

检查结果呈阴性，但事实上是患了病的概率(患病但没有查出来的概率)

$$P(B|\bar{A}) = \frac{P(B)P(\bar{A} | B)}{P(B)P(\bar{A} |B) + P(\bar{B})P(\bar{A}|B)} = \frac{0.0004 * 0.01}{0.0004 * 0.01 + 0.9996 * 0.999} = 4 * 10^{-6}$$
