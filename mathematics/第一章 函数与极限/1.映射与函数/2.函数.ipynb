## 函数

### 1, 函数的概念

定义: 设数集${D}\subset{R}$,则称映射${f}:{D}\rightarrow{R}$为定义在$D$上的函数,

通常简记为$y$=${f({x})}$,${x}\in{D}$

其中$x$称为自变量,y称为因变量,$D$称为定义域，记作$D_{f}$, 即$D_{f}=D$

  函数的定义中,对每个${x}\in{D}$, 按对应法则$f$,总有唯一确定的值$y$与之对应,这个值称为函数$f$在$x$处的函数值,记作$f({x})$,因变量$y$与自变量$x$之间的这种依赖关系,通常称为函数关系,函数值$f({x})$的全体所构成的集合称为函数$f$的值域,记作$R_{f}$或$f({D})$.
  
  即: $R_{f}$=$f({D})$={${y}|{y}=f({x}), {x}\in{D}$}

### 2, 函数的几种特性

1) 函数的有界性

设函数$f({x})$的定义域为$D$,数集$X\subset{D}$,如果存在数$K_1$,

使得$f(x)\leq{k_1}$

对任一$x\in{X}$都成立,那么称函数$f(x)$在${X}$上有上界,而$K_1$称为函数$f(x)$在$X$上的一个上界,如果存在数$K_2$,

使得$f(x) \geq {K_2}$

4) 函数的周期性

设函数$f({x})$的定义域为$D$,如果存在一个正数$l$,使得对于任一$x\in {D}$有$(x\pm{l}\in{D}$, 

且$f(x+{l})=f(x)$

恒成立,那么称$f(x)$为周期函数,$l$称为$f(x)$的周期,通常我们说周期函数的周期是指<font color = red>最小正周期</font>

例如: 函数${sin} \ x$，${cos} \ x$都是以2π为周期的周期函数,函数${tan} \ x$是以π为周期的周期函数。

特殊案例: Dirichlet函数

$D(x)$=$x\in {Q} 1, x\in {Q^C} 0 $

容易验证这是一个周期函数,任何正有理数$r$都是它的周期,因为不存在最小的正有理数,所以它没有最小正周期。


x有理数 + r = 有理数

x无理数 + r = 无理数


### 3, 反函数与复合函数

作为逆映射的特例,有一小反函数的概念

设函数$f:{D}{\rightarrow}f({D})$是单射,则它存在逆映射$f^{-1}:f({D}){\rightarrow}{D}$,称此映射$f{-1}$为函数$f$的反函数。

按此定义,对每个$y\in{f({D})}$, 有唯一的$x\in{D}$使得$f(x)$=$y$,于是有$f^{-1}({y})=x$

这就是说,反函数$f^{-1}$的对应法则是完全由函数$f$的对应法则所确定的。

### 5, 初等函数

五类函数统称为基本初等函数

1. 幂函数:   $y$=$x^u$($u\in{R}$是常熟)
2. 指数函数: $y$=$a^x$(a>0且$a\neq{1}$)
3. 对数函数: $y$ = $log_{a}x$(a>0且$a\neq{1}$,特别当$a$=$e$时,记为$y$=$ln \ x$)
4. 三角函数: $y$ = $sin \ x$, $y$ = $cos \ x$, $y$ = $tan \ x$等,
5. 反三角函数: $y$ = $arcsin \ x$, $y$ = $arccon \ x$ $y$ = $arctan \ x$
